**สำหรับ Activity 3.2 นี้ หัวใจสำคัญคือการฝึกคิดอย่างเป็นขั้นตอน (Algorithmic thinking) โดยใช้ Nested Loops (ลูปซ้อนลูป) ในการควบคุมแถวแนวตั้งและคอลัมน์แนวนอนเพื่อพิมพ์เครื่องหมายดอกจัน * ออกมาเป็นรูปทรงต่างๆ**

หลักการทำงานทั่วไปคือ:
ลูปวงนอก (Outer loop): จะทำหน้าที่ควบคุม "แถว (Row)" หรือแนวตั้ง (เลื่อนจากบนลงล่าง)   

ลูปวงใน (Inner loop): จะทำหน้าที่ควบคุม "คอลัมน์ (Column)" หรือแนวนอน (พิมพ์ตัวอักษรในแถวนั้นๆ จากซ้ายไปขวา)   

**1. รูปสี่เหลี่ยมจัตุรัส (Square of asterisks)**

แนวคิด: สมมติต้องการสี่เหลี่ยมขนาด 5x5 หมายความว่ามีทั้งหมด 5 แถว และในแต่ละแถวจะพิมพ์ดอกจัน 5 ดวงเท่ากันเสมอ   

ลูปวงนอก: วน 5 รอบ (แถวที่ 1 ถึง 5)   

ลูปวงใน: วน 5 รอบคงที่ในทุกๆ แถว เพื่อพิมพ์ *

```python
size = 5
print("--- 1. รูปสี่เหลี่ยมจัตุรัส ---")
for row in range(size):
    for col in range(size):
        print("* ", end="")  # end="" เพื่อให้พิมพ์ต่อกันในแนวนอน
    print()  # เมื่อจบแถว ให้ขึ้นบรรทัดใหม่
```

**2. รูปสามเหลี่ยมมุมฉาก (Right-angled triangle)**

แนวคิด: จำนวนดอกจันในแต่ละแถวจะไม่เท่ากัน แต่จะเพิ่มขึ้นตามเลขแถว เช่น แถวที่ 1 มี 1 ดวง, แถวที่ 2 มี 2 ดวง ไปเรื่อยๆ   

ลูปวงนอก: วนลูปควบคุมแถว (รอบที่ 1 ถึง 5)   

ลูปวงใน: จำนวนรอบจะเปลี่ยนไปตามค่าของลูปวงนอก (ให้วนลูปเท่ากับตัวแปรแถวในรอบนั้นๆ)   

```python
height = 5
print("--- 2. รูปสามเหลี่ยมมุมฉาก ---")
for row in range(1, height + 1):  # แถวที่ 1 ถึง 5
    for col in range(row):        # รอบแรกวน 1 ครั้ง, รอบสองวน 2 ครั้ง...
        print("* ", end="")
    print()  # ขึ้นบรรทัดใหม่เมื่อจบแถว
```

**3. โจทย์ท้าทาย: สามเหลี่ยมมุมฉากแบบกลับหัว (Inverted right-angled triangle)**

แนวคิด: จะตรงข้ามกับข้อ 2 คือ แถวแรกจะมีดอกจันมากที่สุด แล้วค่อยๆ ลดลงทีละ 1 ดวงในแถวถัดไป   

ลูปวงนอก: ใช้ range(height, 0, -1) เพื่อนับถอยหลังตั้งแต่ 5 ลงไปถึง 1  

 ลูปวงใน: วนลูปพิมพ์ดอกจันตามจำนวนรอบที่ลูปวงนอกส่งมา   

```python
height = 5
print("--- 3. รูปสามเหลี่ยมมุมฉากกลับหัว ---")
for row in range(height, 0, -1):  # เริ่มจาก 5 แล้วลดลงทีละ 1 จนถึง 1
    for col in range(row):        # แถวแรกวน 5 ครั้ง, แถวต่อมา 4 ครั้ง...
        print("* ", end="")
    print()
```

**4. โจทย์ท้าทาย: รูปพีระมิด (Pyramid)**

แนวคิด: รูปพีระมิดจะมีความซับซ้อนขึ้น เพราะใน 1 แถว เราต้องจัดการ 2 อย่าง คือ "เว้นวรรค (Spaces)" เพื่อผลักดอกจันไปตรงกลางก่อน แล้วจึงค่อย "พิมพ์ดอกจัน (Asterisks)" ตามไปครับ

    แถวที่ 1: เว้นวรรค 4 ช่อง, พิมพ์ดอกจัน 1 ดวง

    แถวที่ 2: เว้นวรรค 3 ช่อง, พิมพ์ดอกจัน 3 ดวง

    แถวที่ 3: เว้นวรรค 2 ช่อง, พิมพ์ดอกจัน 5 ดวง

    สูตรคณิตศาสตร์ในแต่ละแถว: เว้นวรรค = height - row และ ดอกจัน = (2 * row) - 1

```python
height = 5
print("--- 4. รูปพีระมิด ---")
for row in range(1, height + 1):
    # ลูปวงในตัวที่ 1: พิมพ์ช่องว่างเพื่อจัดกึ่งกลาง
    for space in range(height - row):
        print(" ", end="")
        
    # ลูปวงในตัวที่ 2: พิมพ์ดอกจันเป็นเลขคี่ (1, 3, 5, 7, 9)
    for col in range((2 * row) - 1):
        print("*", end="")
        
    print()  # ขึ้นบรรทัดใหม่
```

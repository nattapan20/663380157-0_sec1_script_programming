หลักการของ Activity 3.1 ตามเอกสาร คือเขาจะมีโค้ดที่ "เขียนผิด" มาให้ 3 ข้อ (Snippet A, B, C)  
สเต็ปนี้กับทุกข้อครับ:  หาจุดบกพร่อง (Bug): ดูว่าบรรทัดไหนเขียนผิดอธิบายสาเหตุ: บอกว่าทำไมมันถึงผิด หรือรันแล้วเกิดอะไรขึ้นแก้ไขโค้ด (Fix): เขียนโค้ดเวอร์ชันที่ถูกต้อง

**ข้อที่ 1: Buggy Snippet A (ปัญหาลูปไม่รู้จบ / Infinite Loop)โค้ดที่มีปัญหา:**

```python
counter = 0
while counter < 5:
    print(counter)
    # Missing counter increment!
```

1. จุดที่เกิดบั๊ก: ภายในบล็อกของ while loop   
2. สาเหตุ: ลืมเพิ่มค่าตัวแปร counter (ไม่มีการเขียน counter += 1) ทำให้ค่า counter เป็น 0 ตลอดเวลา เงื่อนไข 0 < 5 จึงเป็นจริง (True) เสมอ โปรแกรมจะพิมพ์เลข 0 ออกมาทางหน้าจอเรื่อยๆ แบบไม่มีวันหยุดจนเครื่องค้างครับ   

```python
#โค้ดที่แก้ไขให้ถูกต้อง:
counter = 0
while counter < 5:
    print(counter)
    counter += 1  # เพิ่มบรรทัดนี้เพื่อให้อัปเดตค่าขยับไปจนถึง 5 แล้วจบลูป
```

**ข้อที่ 2: Buggy Snippet B (ปัญหาช่วงข้อมูลไม่ครบ / Incorrect Range)โจทย์ตั้งเป้าหมายไว้ว่า: ต้องการพิมพ์เลข 1, 2, 3, 4, 5   โค้ดที่มีปัญหา:**

```python
for i in range(1, 5):
    print(i)
```

1. จุดที่เกิดบั๊ก: ฟังก์ชัน range(1, 5) ในบรรทัดแรก   
2. สาเหตุ: ธรรมชาติของฟังก์ชัน range(start, stop) ใน Python จะนับตั้งแต่ค่า start ไปจนถึงค่า ก่อนหน้า stop (ไม่รวมตัวหยุด) ดังนั้น range(1, 5) จะพิมพ์ออกมาแค่เลข 1, 2, 3, 4 ซึ่งไม่ครบตามเป้าหมายที่อยากได้เลข 5 ด้วย   

```python
# โค้ดที่แก้ไขให้ถูกต้อง: เปลี่ยนตัวหยุด (stop) จาก 5 เป็น 6 เพื่อให้นับถึง 5
for i in range(1, 6):
    print(i)
```

**ข้อที่ 3: Buggy Snippet C (ปัญหาการวาง Break ผิดตำแหน่ง / Misplaced Break)โจทย์ตั้งเป้าหมายไว้ว่า: ต้องการพิมพ์หาตัวอักษร 'l' ในคำว่า "hello"   โค้ดที่มีปัญหา:**

```python
for char in "hello":
    if char == 'l':
        print("Found 'l', stopping!")
        break
    print(char)  # บรรทัดนี้คือปัญหา
```

1. จุดที่เกิดบั๊ก: การจัดย่อหน้า (Indentation) ของคำสั่ง print(char) ที่อยู่บรรทัดสุดท้าย   
2. สาเหตุ: ตัวคำสั่ง print(char) ดันไปอยู่นอกเงื่อนไข if แต่อยู่ใน for loop หลัก ทำให้มันจะทำงานทุกครั้งที่วนลูป ผลลัพธ์ที่ได้เลยกลายเป็นพิมพ์ 'h', พิมพ์ 'e', พิมพ์ 'l' (จากในเงื่อนไขหลุดออกมาพิมพ์ข้างนอกอีกรอบ) แล้วค่อยสั่ง break   
3. โค้ดที่แก้ไขให้ถูกต้อง:ถ้าต้องการให้โปรแกรมแสดงเฉพาะตัวอักษรก่อนหน้าที่จะเจอ 'l' และไม่พิมพ์ 'l' ซ้ำซ้อน ต้องปรับให้ print(char) ทำงานเฉพาะตอนที่เงื่อนไข if เป็นเท็จเท่านั้น (ใช้ else เข้าช่วย หรือจัดตำแหน่งให้ถูกต้อง)

```python
# โค้ดที่แก้ไขให้ถูกต้อง:
for char in "hello":
    if char == 'l':
        print("Found 'l', stopping!")
        break
    else:
        print(char)  # ขยับเข้ามาอยู่ใน else เพื่อพิมพ์เฉพาะตัวที่ไม่ใช่ 'l'
```

ACTIVITY 2.2
Choose Your Own Adventure

ชื่อเรื่อง การผจญภัยตามล่าสมบัติในป่าลึกลับ

```python
# ACTIVITY 2.2
# Choose Your Own Adventure

print("=== Welcome to the Adventure Game ===")
print("You are an explorer searching for a hidden treasure in a mysterious forest.")

# Decision Point 1
choice = input("You arrive at a crossroads. Do you go left or right? (left/right): ").lower()

if choice == "left":
    print("\nYou discover a dark cave.")

    # Decision Point 2
    choice = input("Do you enter the cave or walk away? (enter/leave): ").lower()

    if choice == "enter":
        print("\nYou find a treasure chest.")

        # Decision Point 3
        choice = input("Do you open the chest? (yes/no): ").lower()

        if choice == "yes":
            print("\n🎉 You found the hidden treasure! You Win!")
        elif choice == "no":
            print("\nYou leave safely but without any treasure.")
        else:
            print("\nInvalid choice. The adventure ends.")

    elif choice == "leave":
        print("\nYou leave the cave and safely return home.")
    else:
        print("\nInvalid choice. The adventure ends.")

elif choice == "right":
    print("\nYou meet a hungry wolf.")

    # Decision Point 2
    choice = input("Do you fight or run? (fight/run): ").lower()

    if choice == "fight":
        print("\nThe wolf defeats you. Game Over.")
    elif choice == "run":
        print("\nYou escape safely but never find the treasure.")
    else:
        print("\nInvalid choice. The adventure ends.")

else:
    print("\nInvalid choice. The adventure ends.")

print("\n=== End of Game ===")
```

ชื่อเรื่อง

    การผจญภัยตามล่าสมบัติในป่าลึกลับ

เนื้อเรื่อง (Story)

    ผู้เล่นเป็นนักสำรวจที่ต้องการค้นหาสมบัติที่ซ่อนอยู่ในป่าลึกลับ ระหว่างทางผู้เล่นจะต้องตัดสินใจหลายครั้ง ซึ่งแต่ละการตัดสินใจจะนำไปสู่ผลลัพธ์ที่แตกต่างกันเช่น ได้สมบัติ กลับบ้านอย่างปลอดภัย หรือเกมจบ (Game Over)

จุดตัดสินใจ (Decision Points)

จุดตัดสินใจที่ 1

โปรแกรมถามผู้เล่น
คุณต้องการไปทางซ้ายหรือทางขวา?
(left/right)
ถ้าเลือก
left → พบถ้ำลึกลับ
right → พบหมาป่าหิวโซ


จุดตัดสินใจที่ 2
กรณีเลือกทางซ้าย

ถามผู้เล่น
คุณต้องการเข้าถ้ำหรือกลับ?
(enter/leave)
ถ้า
enter → พบหีบสมบัติ
leave → กลับบ้านอย่างปลอดภัย

กรณีเลือกทางขวา

ถามผู้เล่น
คุณจะสู้หรือหนี?
(fight/run)
ถ้า
fight → หมาป่าชนะ เกมจบ
run → หนีรอด แต่ไม่ได้สมบัติ


จุดตัดสินใจที่ 3

(เกิดขึ้นเมื่อผู้เล่นเข้าถ้ำ)

ถามผู้เล่น
คุณต้องการเปิดหีบสมบัติหรือไม่?
(yes/no)
ถ้า
yes → ได้สมบัติและชนะเกม
no → กลับบ้านอย่างปลอดภัย แต่ไม่ได้สมบัติ

ลำดับการทำงาน (Pseudo-code)
เริ่มเกม

แสดงข้อความต้อนรับ

รับข้อมูลจากผู้เล่น (left/right)

if เลือก left

    รับข้อมูล (enter/leave)

    if enter

        รับข้อมูล (yes/no)

        if yes

            แสดง "คุณพบสมบัติ! คุณชนะ"

        else

            แสดง "คุณกลับบ้านอย่างปลอดภัย แต่ไม่ได้สมบัติ"

    else

        แสดง "คุณกลับบ้านอย่างปลอดภัย"

elif เลือก right

    รับข้อมูล (fight/run)

    if fight

        แสดง "หมาป่าชนะ เกมจบ"

    else

        แสดง "คุณหนีรอด แต่ไม่ได้สมบัติ"

else

    แสดง "เลือกคำตอบไม่ถูกต้อง"

อธิบายการใช้ input()

เกมนี้ใช้ input() ทั้งหมด 3 ครั้ง

รับข้อมูลว่าจะไปทางซ้ายหรือขวา
รับข้อมูลว่าจะเข้าถ้ำหรือกลับ (หรือสู้หรือหนี)
รับข้อมูลว่าจะเปิดหีบหรือไม่

อธิบายการใช้ if-elif-else

ใช้ if-elif-else เพื่อให้โปรแกรมตรวจสอบคำตอบของผู้เล่น

ถ้าผู้เล่นเลือก left โปรแกรมจะไปยังเหตุการณ์ในถ้ำ
ถ้าผู้เล่นเลือก right โปรแกรมจะไปยังเหตุการณ์หมาป่า
ถ้าผู้เล่นพิมพ์อย่างอื่น โปรแกรมจะแจ้งว่าเลือกไม่ถูกต้อง

ภายในแต่ละเส้นทางยังใช้ if ซ้อน (Nested if) เพื่อสร้างจุดตัดสินใจเพิ่มเติม ทำให้เกมมีหลายตอนจบ

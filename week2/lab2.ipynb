Lab 2.1: Number Classifier ○ Objective: Students will write a script to classify a user-provided integer as positive, negative, or zero, and also check if it's even or odd.

```python
# Lab 2.1:
num = int(input("Enter a number: "))

if num < 0:
    text = "negative"
elif num > 0:
    text = "positive"
else:
    text = "zero"

if num % 2 == 0:
    even_odd = "even"
else:
    even_odd = "odd"

print(f"{num} The number is {text} and {even_odd}.")
```

Lab 2.2: Simple Decision-Making Program (Movie Rating) ○ Objective: Students will create a program that recommends a movie rating based on the user's age, demonstrating nested or chained conditionals. ○ Tasks (Common for Local & Colab):

    Create a new Python file (lab2_2.py) or Colab code cell.
    Prompt the user to enter their age. Convert it to an integer.
    Implement the following logic: ■ If age is less than 5: Print "You're too young for movies! Enjoy cartoons." ■ If age is 5 to 12 (inclusive): Recommend "G-rated or PG-rated movies." ■ If age is 13 to 17 (inclusive): Recommend "PG-13 or R-rated (with parental guidance)." ■ If age is 18 or older: Recommend "Any movie rating."
    Challenge: Add a secondary prompt: "Do you like action movies? (yes/no)". If the age is 18 or older AND they like action, print "You might enjoy the latest action blockbuster!" (This requires a nested if or more complex logical and conditions).


```python
# Lab 2.2:
age = int(input("Enter your age: "))

if age < 5:
    print("You're too young for movies! Enjoy cartoons.")
elif age <= 12:
    print("G-rated or PG-rated movies.")
elif age <= 17:
    print("PG-13 or R-rated (with parental guidance).")
else:
    print("Any movie rating.")

action = input("Do you like action movies? (yes/no): ").lower()

if age >= 18 and action == "yes":
    print("You might enjoy the latest action blockbuster!")
```
